```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image1=cv2.imread ('bild.jpg') 
gray_image = cv2.cvtColor(image1,cv2.COLOR_BGR2GRAY)
cv2.imshow('GRAY IMAGE',gray_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
edges1 = cv2.Canny(gray_image, 120, 150)
plt.imshow(edges1,cmap='gray')
plt.title('canny_edges')
plt.show()
```

```python
lines=cv2.HoughLinesP(edges1,1,np.pi/180,threshold=300,minLineLength=200,maxLineGap=200)
plt.imshow(lines)
plt.title('Detecting the points')
plt.show
```

```python
for line in lines:
    x1,y1,x2,y2=line[0]
    cv2.line(image1,(x1,y1),(x2,y2),(255,0,0),3)
    plt.imshow(image1)
plt.title('Lines')
plt.show
```
